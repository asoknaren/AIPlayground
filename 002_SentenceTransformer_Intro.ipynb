```python
from sentence_transformers import SentenceTransformer
```

### Load the SentenceTransformer model

```python
model = SentenceTransformer("sentence-transformers/all-mpnet-base-v2")
```

### Encode a sentence

```python
sentences = ["This is an example sentence", "Each sentence is converted."]
embeddings = model.encode(sentences)
print("Embeddings for the first sentence:")
print(embeddings[0])
print(f"\nShape of embeddings: {embeddings.shape}")
```

```python
print(f"The size (shape) of the embeddings is: {embeddings.shape}")
```

### Generate tokens for a given text

```python
text_to_tokenize = "Hello, how are you today?"

# The SentenceTransformer model has a tokenizer as part of its internal structure.
# You can access it via model.tokenizer
tokens = model.tokenizer.tokenize(text_to_tokenize)

print(f"Original text: {text_to_tokenize}")
print(f"Tokens: {tokens}")
```
